```python
import pandas as pd
import numpy as np
```

```python
# title.basics.tsv.gz url
basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
# load basic URL
basics = pd.read_csv(basics_url, sep='\t', low_memory = False)
```

```python
basics.head()
```

```python
# title.ratings.tsv.gz
ratings_url = "https://datasets.imdbws.com/title.ratings.tsv.gz"
# Load raitings URL
ratings = pd.read_csv(ratings_url, sep='\t', low_memory = False)
```

```python
ratings.head()
```

```python
# title.akas.tsv.gz
akas_url = "https://datasets.imdbws.com/title.akas.tsv.gz"
# Load akas URL
akas = pd.read_csv(akas_url, sep='\t', low_memory = False)
```

```python
akas.head()
```

## Replacing \N with nan

```python
# Replacing \N in basics
basics.replace({"\\N":np.nan}, inplace = True)
```

```python
# Confirm Replacement 
basics.head()
```

```python
# Replace \N in ratings
ratings.replace({"\\N":np.nan}, inplace = True)
```

```python
#Replace \N in akas
akas.replace({"\\N":np.nan}, inplace = True)
```

```python
# Confirm replacement 
akas.head()
```

## Basics Processing

```python
basics.head()
```

```python
# Eliminate movies that are null for runtimeMinutes 
basics.dropna(subset = ["runtimeMinutes"], inplace = True)
```

```python
# Confirm drop
basics["runtimeMinutes"].isnull().sum()
```

```python
# Eliminate movies that are null for genre
basics.dropna(subset = ["genres"], inplace = True)
```

```python
# Confirm Drop 
basics["genres"].isnull().sum()
```

```python
basics["titleType"].value_counts()
```

```python
# keep only titleType==Movie
basics = basics.loc[(basics["titleType"]=="movie")]
```

```python
# Confirm 
basics.head()
```

```python
# keep startYear 2000-2022
basics = basics.loc[(basics["startYear"]>"1999")&(basics["startYear"]<"2022")]
```

```python
# Confirm
basics.head()
```

```python
# Eliminate movies that include  "Documentary" in genre (see tip below)
is_doc = basics["genres"].str.contains("documentary", case = False)
basics = basics[~is_doc]
```

```python
#Confirm
basics.head()
```

## AKAS Processing

```python
akas.head()
```

```python
# keep only US entries
akas = akas.loc[(akas["region"]== "US")]
```

```python
# Confirm
akas.head()
```

## Filtering dataframes

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
keepers = basics["tconst"].isin(akas["titleId"])
keepers
```

```python
# filter basics
basics = basics[keepers]
basics
```

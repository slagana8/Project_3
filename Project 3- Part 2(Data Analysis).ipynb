```python
# Imports 
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
# Load 2000 dataframe
two_thousand = pd.read_csv("Data/final_tmdb_data2000.csv.gz")
two_thousand.head()
```

```python
two_thousand.info()
```

```python
# Load 2001 df
two_thousand_one = pd.read_csv("Data/final_tmdb_data2001.csv.gz")
two_thousand_one.head()
```

```python
two_thousand_one.info()
```

```python
# Concatenate the data into one dataframe
final_df = pd.concat([two_thousand,two_thousand_one])
final_df.head()
```

```python
final_df.info()
```

Q1. How many movies had at least some valid financial information (values > 0 for budget OR revenue)?

```python
final_df = final_df.dropna(subset = ["budget", "revenue"])
```

```python
financial_df = final_df[(final_df["budget"]>0) | (final_df["revenue"]>0)]
```

```python
print("The amount of movies that had budgets or revenue greater than 0:",len(financial_df.index))
```

Q2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
# Value counts of each certification
financial_df["certification"].value_counts()
```

```python
# Drop films with NR rating
financial_df= financial_df[financial_df.certification != "NR"]
```

```python
# Value Counts of certification categories (G/PG/PG-13/R)
financial_df["certification"].value_counts()
```

```python
# Visual of Certification Categories
ax=sns.countplot(x=financial_df['certification'])
ax.set_title('Films per Certification Category');
```

Q3. What is the average revenue per certification category?

```python
# Average revenue per cert category 
financial_df.groupby("certification")["revenue"].mean()
```

```python
# Visual of Average revenue per cert category
ax=sns.barplot(x=financial_df['certification'], y=financial_df['revenue'])
ax.set_title('Average Revenue per Certfication Category');
```

Q4. What is the average budget per certification category?

```python
# Average budget per cert category 
financial_df.groupby("certification")["budget"].mean()
```

```python
# Visual of Average revenue per cert category
ax=sns.barplot(x=financial_df['certification'], y=financial_df['budget'])
ax.set_title('Average Budget per Certfication Category');
```

```python
# Saving final tmdb results 
tmdb_results_combined = final_df
tmdb_results_combined.to_csv("Data/tmdb_results_combined.csv.gz", compression = "gzip", index = False)
```
